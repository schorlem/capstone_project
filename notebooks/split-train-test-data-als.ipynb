```python
import os
from IPython.display import display
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/home/andre/Documents/mooc/udacity/machinelearning/capstone_project/data/questions.csv")
```

```python
display(data.head())

print "Rows in data: {}".format(len(data))
print "Ratio of duplicates: {:.2f}".format(float(len(data[data.is_duplicate == 1]))/len(data))
```

```python
labels = data.iloc[:, -1].values
# Stratified splitting is enabled because the data is somewhat imbalanced(~37% are duplicates)
df_train, df_test = train_test_split(data, test_size=0.1, stratify=labels, random_state=42)
```

```python
display(df_train.head())
display(df_test.head())

print "Rows in test set: {} and train set: {}".format(len(df_test), len(df_train))
print "Ratio: {:.2f}".format(float(len(df_test))/len(df_train))
print "Ratio of duplicates test: {:.2f}".format(float(len(df_test[df_test.is_duplicate == 1]))/len(df_test))
print "Ratio of duplicates train: {:.2f}".format(float(len(df_train[df_train.is_duplicate == 1]))/len(df_train))
```

```python
output_dir = "../output/data/"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df_test.to_pickle(output_dir+"test_data.pkl")
df_train.to_pickle(output_dir+"train_data.pkl")
```
